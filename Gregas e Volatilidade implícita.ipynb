```python
# Função para calcular o preço da opção usando a formula de Black-Scholes
def black_scholes_price(option_type, K, r, q, T, sigma, S):
    d1 = (np.log(S / K) + (r - q + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == "call":
        return S * np.exp(-q * T) * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == "put":
        return K * np.exp(-r * T) * norm.cdf(-d2) - S * np.exp(-q * T) * norm.cdf(-d1)
    else:
        raise ValueError("Option type must be either 'call' or 'put'.")

# Cálculo das gregas
def calculate_greeks(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))
    vega = S * norm.pdf(d1) * np.sqrt(T)

    if option_type == 'call':
        delta = norm.cdf(d1)
        theta = (-S * norm.pdf(d1) * sigma / (2 * np.sqrt(T)) -
                 r * K * np.exp(-r * T) * norm.cdf(d2))
        rho = K * T * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        delta = -norm.cdf(-d1)
        theta = (-S * norm.pdf(d1) * sigma / (2 * np.sqrt(T)) +
                 r * K * np.exp(-r * T) * norm.cdf(-d2))
        rho = -K * T * np.exp(-r * T) * norm.cdf(-d2)
    else:
        raise ValueError("option_type deve ser 'call' ou 'put'")

    return {'Delta': delta, 'Gamma': gamma, 'Vega': vega / 100,
            'Theta': theta / 365, 'Rho': rho / 100}

# Exemplo de uso
rf =  float(mark_price[0]['riskFreeInterest'])
S = df['Close'].iloc[-1]         # Preço atual do BTC
K = 83000          # Strike price 
T = 1/365          # Tempo até vencimento em anos 
r = rf             # Taxa de juros anualizada
sigma = np.std(df['return'].dropna()) * np.sqrt(365)       # Volatilidade anualizada
q = 0

# Calcular preço da Call e Put
call_price = black_scholes_price('call', K, r, q, T, sigma, S)
put_price = black_scholes_price('put', K, r, q, T, sigma, S)

# Calcular gregas
call_greeks = calculate_greeks(S, K, T, r, sigma, 'call')
put_greeks = calculate_greeks(S, K, T, r, sigma, 'put')

print(f"Preço da Call: {call_price:.2f} USD")
print(f'''Gregas da Call: 
    Delta: {call_greeks['Delta']}
    Gamma: {call_greeks['Gamma']}
    Vega: {call_greeks['Vega']}
    Theta: {call_greeks['Theta']}
    Rho: {call_greeks['Rho']}    
''')

print(f"Preço da Put: {put_price:.2f} USD")
print(f'''Gregas da Put: 
    Delta: {put_greeks['Delta']}
    Gamma: {put_greeks['Gamma']}
    Vega: {put_greeks['Vega']}
    Theta: {put_greeks['Theta']}
    Rho: {put_greeks['Rho']}    
''')
```

```python
# Parâmetros iniciais:
passo = 1e-8                                   # Passo para o cálculo da derivada aproximada
sigma_initial = np.std(df['return'].dropna()) * np.sqrt(365)  # Chute inicial para sigma
S = df['Close'].iloc[-1]                       # Preço do ativo
r = float(mark_price[0]['riskFreeInterest'])   # Taxa de juros livre de risco
T = 1 / 365                                    # Prazo da opção (vencimento)
strikes = range(80000, 90001, 1000)

# Coeficientes para a aproximação da função de distribuição normal acumulada
gama = 0.2316419
a1 = 0.319381530
a2 = -0.356563782
a3 = 1.781477937
a4 = -1.821255978
a5 = 1.330274429

# Dicionário para armazenar a volatilidade implícita para cada strike
implied_vols = {}

# Loop para strikes de 80000 a 90000 (passo de 1000, por exemplo)
for X in strikes:
    # Obtém o preço de mercado da opção para o strike X
    value = client.options_mark_price(symbol=f"BTC-250407-{X}-C")
    C = float(value[0]['markPrice'])
    
    # Inicializa a lista de sigma para o método de Newton-Raphson
    sigma_list = [sigma_initial]
    
    # Loop de iterações (por exemplo, 100 iterações)
    for i in range(1000):
        sigma_val = sigma_list[i]
        
        # Cálculo de d1 e d2
        d1 = (np.log(S/X) + (r + sigma_val**2/2) * T) / (sigma_val * np.sqrt(T))
        d2 = d1 - sigma_val * np.sqrt(T)
        
        # Cálculo das densidades da normal
        Nlinhad1 = 1 / (np.sqrt(2*np.pi)) * np.exp(-d1**2/2)
        Nlinhad2 = 1 / (np.sqrt(2*np.pi)) * np.exp(-d2**2/2)
        
        # Cálculo dos fatores kd1 e kd2
        kd1 = 1 / (1 + gama * d1)
        kd2 = 1 / (1 + gama * d2)
        
        # Aproximação de Nd1
        if d1 >= 0:
            Nd1 = 1 - Nlinhad1 * (a1*kd1 + a2*kd1**2 + a3*kd1**3 + a4*kd1**4 + a5*kd1**5)
        else:
            d1_temp = -d1
            kd1 = 1 / (1 + gama * d1_temp)
            Nlinhad1 = 1 / (np.sqrt(2*np.pi)) * np.exp(-d1_temp**2/2)
            Nd1 = 1 - Nlinhad1 * (a1*kd1 + a2*kd1**2 + a3*kd1**3 + a4*kd1**4 + a5*kd1**5)
            Nd1 = 1 - Nd1
        
        # Aproximação de Nd2
        if d2 >= 0:
            Nd2 = 1 - Nlinhad2 * (a1*kd2 + a2*kd2**2 + a3*kd2**3 + a4*kd2**4 + a5*kd2**5)
        else:
            d2_temp = -d2
            kd2 = 1 / (1 + gama * d2_temp)
            Nlinhad2 = 1 / (np.sqrt(2*np.pi)) * np.exp(-d2_temp**2/2)
            Nd2 = 1 - Nlinhad2 * (a1*kd2 + a2*kd2**2 + a3*kd2**3 + a4*kd2**4 + a5*kd2**5)
            Nd2 = 1 - Nd2
        
        # f: diferença entre o preço teórico e o preço de mercado
        f_val = S * Nd1 - X * np.exp(-r * T) * Nd2 - C
        
        # Cálculo da derivada por diferença finita
        sigma_aux = sigma_val + passo
        d1_aux = (np.log(S/X) + (r + sigma_aux**2/2) * T) / (sigma_aux * np.sqrt(T))
        d2_aux = d1_aux - sigma_aux * np.sqrt(T)
        
        Nlinhad1_aux = 1 / (np.sqrt(2*np.pi)) * np.exp(-d1_aux**2/2)
        Nlinhad2_aux = 1 / (np.sqrt(2*np.pi)) * np.exp(-d2_aux**2/2)
        
        kd1_aux = 1 / (1 + gama * d1_aux)
        kd2_aux = 1 / (1 + gama * d2_aux)
        
        if d1_aux >= 0:
            Nd1_aux = 1 - Nlinhad1_aux * (a1*kd1_aux + a2*kd1_aux**2 + a3*kd1_aux**3 + a4*kd1_aux**4 + a5*kd1_aux**5)
        else:
            d1_aux_temp = -d1_aux
            kd1_aux = 1 / (1 + gama * d1_aux_temp)
            Nlinhad1_aux = 1 / (np.sqrt(2*np.pi)) * np.exp(-d1_aux_temp**2/2)
            Nd1_aux = 1 - Nlinhad1_aux * (a1*kd1_aux + a2*kd1_aux**2 + a3*kd1_aux**3 + a4*kd1_aux**4 + a5*kd1_aux**5)
            Nd1_aux = 1 - Nd1_aux
        
        if d2_aux >= 0:
            Nd2_aux = 1 - Nlinhad2_aux * (a1*kd2_aux + a2*kd2_aux**2 + a3*kd2_aux**3 + a4*kd2_aux**4 + a5*kd2_aux**5)
        else:
            d2_aux_temp = -d2_aux
            kd2_aux = 1 / (1 + gama * d2_aux_temp)
            Nlinhad2_aux = 1 / (np.sqrt(2*np.pi)) * np.exp(-d2_aux_temp**2/2)
            Nd2_aux = 1 - Nlinhad2_aux * (a1*kd2_aux + a2*kd2_aux**2 + a3*kd2_aux**3 + a4*kd2_aux**4 + a5*kd2_aux**5)
            Nd2_aux = 1 - Nd2_aux
        
        faux = S * Nd1_aux - X * np.exp(-r * T) * Nd2_aux - C
        derivf = (faux - f_val) / (sigma_aux - sigma_val)
        
        # Atualiza sigma com o método de Newton-Raphson
        sigma_new = sigma_val - f_val / derivf
        sigma_list.append(sigma_new)
    
    # Armazena a volatilidade implícita final para o strike atual
    implied_vols[X] = sigma_list[-1]

# Exibe os resultados: volatilidade implícita para cada strike
print("Strike\tVolatilidade Implícita")
for strike, vol in implied_vols.items():
    print(f"{strike}\t{vol}")
```

```python
def trajetorias_GBM(S0, mu, sigma, T, n_steps, n_trajs):

    # criando um array com os tempos
    delta_t = T / n_steps
    tempo = np.arange(start = 0, stop = T + 0.9 * delta_t, step = delta_t)


    # calculando média e desv_pad dos retornos
    media = mu * delta_t
    desv_pad = sigma * np.sqrt(delta_t)

    # criando um array para os preços do underlying em cada trajetória
    St = np.zeros((n_steps + 1, n_trajs))
    St_ant = np.zeros((n_steps + 1, n_trajs))


    # determinando o valor St = S0 para t = 0 em todas as trajetórias
    St[0, :] = S0
    St_ant[0, :] = S0


    # gerando o retornos do underlying com distribuição normal N(media, desv_pad) para cada trajetória
    eps = np.random.normal(0, 1, (n_steps, n_trajs))
    ret = media + desv_pad * eps
    ret_ant = media + desv_pad * (-eps)


    # For loop para os n passos da simulação
    for i in range(1, n_steps + 1):
        St[i, :] = St[i-1, :] * (1 + ret[i-1, :])
        St_ant[i, :] = St_ant[i-1, :] * (1 + ret_ant[i-1, :])


    St = np.hstack((St_ant, St))

    # criando um DataFrame com cada trajetória de preços do underlying, e com o array de tempo como index
    df = pd.DataFrame(St, index = tempo)

    return df

# simulando trajetórias do preço do ativo com vol implícita
simul01 = trajetorias_GBM(S0 = df['Close'].iloc[-1], mu =np.mean(df['return'].dropna()) * 365, sigma = vol_obs, T = 1/365, n_steps = 1*(365), n_trajs = 1000)

# selecionando apenas o último preço de cada trajetória: S_T
sT01 = simul01.iloc[-1, :]

print(f'A Vol implícita é: {vol_obs}')

# ajustando o tamanho da figura
figure(figsize = (20, 8))

# plotando as trajetórias
plt.subplot(1, 10, (1, 7))
plt.plot(simul01);


# plotando os resultados
plt.subplot(1, 10, (8, 10))
plt.hist(sT01, bins = 35, alpha = 0.5, orientation = 'horizontal');
# plt.xlim(xmin = 250, xmax = 50)
```
